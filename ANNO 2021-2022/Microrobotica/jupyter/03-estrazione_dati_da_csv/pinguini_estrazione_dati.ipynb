Struttura dizionario: 

dizionario = {"species":[], "island":[] ...

```python
dict_dati = {}
```

```python
with open("./penguins_size.csv", "r") as file:
    for nome_colonna in file.readlines()[0][:-1].split(","): # il [:-1] lo uso per escludere lo \n alla fine
        dict_dati[nome_colonna] = []
    
dict_dati
```

```python
with open("./penguins_size.csv", "r") as file:
    for riga in file.readlines()[1:]:
        dati_riga = riga[:-1].split(",")
        if "NA" not in dati_riga:
            dict_dati["species"].append(dati_riga[0])
            dict_dati["island"].append(dati_riga[1])
            dict_dati["culmen_length_mm"].append(float(dati_riga[2]))
            dict_dati["culmen_depth_mm"].append(float(dati_riga[3]))
            dict_dati["flipper_length_mm"].append(float(dati_riga[4]))
            dict_dati["body_mass_g"].append(dati_riga[5])
            dict_dati["sex"].append(dati_riga[6])
        
dict_dati
```
