### MATPLOTLIB

Permette di creare grafici anche molto complessi.

Prende in input array numpy.

Esistono due modi per creare grafici, **pyplot** e un'altra **orientata agli oggetti (OO)**.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.random.uniform(0,100, 100)
a
```

```python
# grafico in modalit OO

fig, ax = plt.subplots(figsize=(10,10))
ax.plot(a) # io con ax vado a moficare l'asse
# se io ad ax passo un array numpy mette sull'asse x l'indice degli elementi dentro l'array e sull'asse y il valore
plt.show()
```

```python
x = np.linspace(0,2*np.pi, 100) # genera 100 punti equispaziati tra 0 e 2
print(x)

y = np.sin(x)
print(y)

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,2*np.pi, 100) # genera 100 punti equispaziati tra 0 e 2
print(x)

y = 1/x

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(x,y)
plt.show()
```

```python
date = np.array(["01/01/2021", "02/01/2021", "03/01/2021", "04/01/2021"])
valori = np.array([3.4,5.1,1.3,2.8])
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(date, valori)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(date, valori, "r")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(date, valori, "r-")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(date, valori, "r--")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(date, valori, "ro")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(date, valori, "o")
plt.show()
```

```python
n_lanci = 100

dado1 = np.random.randint(1,7,n_lanci)
dado2 = np.random.randint(1,7,n_lanci)

somma = dado1 + dado2

fig, ax = plt.subplots(figsize=(6,4))
ax.hist(dado1, bins=6) # bins equivale al numero di combinazioni diverse che possono uscire dal lancio dei dadi
# nell'istogramma equivale al numero di barre
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.hist(dado2, bins=6)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.hist(somma, bins=11)
plt.show()
```
