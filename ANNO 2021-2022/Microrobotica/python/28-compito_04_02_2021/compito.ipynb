```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
"""
Scarica il file marziani.csv e crea un notebook in cui riproduci il grafico visto a lezione, 
per una specie marziana a tua scelta.
"""
```

```python
marziani = pd.read_csv("./marziani.csv")
marziani
```

```python
marziani.info()
```

```python
marziani_non_NAN = marziani.dropna()
marziani_non_NAN
```

```python
marziani_non_NAN.info()
```

```python
nomi_colonne = np.delete(marziani_non_NAN.keys(), 0)
marziani_scelti = marziani_non_NAN[(marziani_non_NAN["specie"] == "Robby")]

fig, axs = plt.subplots(nrows=5, ncols=5, figsize=(30,30))

for x, nom_col_x in enumerate(nomi_colonne):
    for y, nom_col_y in enumerate(nomi_colonne):
        if nom_col_y == nom_col_x:
            axs[x,y].hist(marziani_scelti[nom_col_y])
            axs[x,y].set_xlabel(nom_col_y)
            axs[x,y].set_title(f"{nom_col_y}")
        else:
            axs[x,y].scatter(marziani_scelti[nom_col_y],marziani_scelti[nom_col_x])
            axs[x,y].set_xlabel(nom_col_x)
            axs[x,y].set_ylabel(nom_col_y)
            
            axs[x,y].set_title(f"{nom_col_x}/{nom_col_y}")
        

plt.show()
```
