1)Creare una funzione che trovi gli elementi comuni a due ndarray monodimensionali di lunghezza diversa e gli elimini.

```python
array1 = np.random.randint(0,100,20)
array2 = np.random.randint(0,100,10)

print(array1)
print(array2)
```

```python
difference = set(array1).symmetric_difference(set(array2)) # faccio la differenza simmetrica 
list_difference = list(difference)
list_difference = np.array(list_difference) # transformo in array numpy
list_difference
```

2)Usiamo Numpy per creare dei dati simulati riferiti ad una situazione reale: in certi casi i dati simulati sono una buona approssimazione della realt. Nel parco vicino alla scuola  installata una centralina per la misurazione del livello di polveri sottili PM10 nell'aria. Talvolta la centralina va in uno stato di errore per cui la misurazione delle PM10 fallisce ed il dato corrispondente riporta il valore 9999.9 all'interno dell'array delle misurazioni dati. Creare un array di 1000 campioni che generino dei valori secondo una distribuzione normale con valore medio 19 e varianza 5 e supporre che la centralina sia andata in errore 20 volte. La distribuzione gaussiana  una buona approssimazione dei dati reali della centralina? 

```python
varianza = 5
valore_medio = 19
numero_campioni = 1000
dev_str = np.sqrt(varianza)
campioni = np.random.normal(valore_medio, dev_str,1000)
campioni
```

```python
array_indici_dati_sbagliati = np.random.choice(range(numero_campioni), 20, replace=False)
# io con replace=False indico che gli indici gi estratti non deve estrairli di nuovo
array_indici_dati_sbagliati
```

```python
campioni[array_indici_dati_sbagliati] = 9999.9
campioni
```
