```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(0,1,100)
y = np.random.uniform(0,1,100)
```

```python
# SCATTER

fig, ax = plt.subplots(figsize =(10,10))
ax.scatter(x,y)
plt.show()
```

```python
# PLOT

fig, ax = plt.subplots(figsize =(10,10))
ax.plot(x,y, "o")
plt.show()
```

```python
# CON PLOT PER NON POSSO MODIFICARE I SINGOLI PUNTI

# POSSO MODIFICARE LA SIZE
size = np.random.uniform(100,10,100)

fig, ax = plt.subplots(figsize =(10,10))
ax.scatter(x,y, s=size)
plt.title("MODIFICO SIZE")
plt.show()

# POSSO MODIFICARE IL COLORE
col = np.random.choice(["r", "g", "b", "y"], 100)

fig, ax = plt.subplots(figsize =(10,10))
ax.scatter(x,y, c=col)
plt.title("MODIFICO COLORE")
plt.show()

# POSSO MODIFICARE SIA SIZE CHE COLORE
col = np.random.choice(["r", "g", "b", "y"], 100)

fig, ax = plt.subplots(figsize =(10,10))
ax.scatter(x,y, c=col, s=size)
plt.title("MODIFICO SIA SIZE CHE COLORE")
plt.show()

# POSSO MODIFICARE SIA SIZE CHE COLORE CHE TRASPARENZA
col = np.random.choice(["r", "g", "b", "y"], 100)

fig, ax = plt.subplots(figsize =(10,10))
ax.scatter(x,y, c=col, s=size, alpha=0.2)
plt.title("POSSO MODIFICARE SIA SIZE CHE COLORE CHE TRASPARENZA")
plt.show()
```

```python
# ISTOGRAMMA
"""
Gli istogrammi servono per rappresentare la frenquenza con cui avvengono gli eventi.

Hist capisce i possbili valori che sono contenuti dentro una lista  in automatico gli interpreta.
"""

fig, ax = plt.subplots(figsize=(10,10))
ax.hist(col)
plt.show()
```

```python
# RAPPRESENTAZIONE MATRICI

"""
Ci sono 2 modi per rappresnetare le matrici e di conseguenza le immagini:
- colori: contengono 3 valori per pixel (red,green,blue)
- scala di grigio: massima intensit 1 e colore bianco, minima intensit 0 e colore nero

Solitamente i valori vanno da 0 a 255 ma possono anche andare da 0 a 1 o anche altri range.
"""

matrice = np.random.uniform(0,1,(100,100))

fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(matrice)
plt.show()

"""
Matplotlib adotta una mappa di colore per colorarla di default, in teoria sarebbe in scala di grigi.
"""

fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(matrice, cmap="Greys")
plt.show()

"""
Matplotlib trova il pi piccolo e grande valore ed in seguito mappa gli altri valori in base a questi.
Si possono anche forzare il massimo e il minimo valore.
I valori fuori range prenderanno il valori di minimo o di massimo.
"""

fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(matrice, cmap="Greys", vmax= 0.7, vmin=0.3)
plt.show()
```

```python
# MANIPOLAZIONE MATRICI

matrice1 = np.copy(matrice)
matrice1[:, 10:20] = 0
fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(matrice1, cmap="Greys")
plt.show()
```

```python
# LETTURA IMMAGINI

imm = plt.imread("./gatto.jpg")
imm
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(imm)
plt.show()
```

```python
imm1 = np.copy(imm)
imm1[40:80, 40:80] = np.array([0,255,123])

# ATTENZIONE: nel grafico la x  uguale al numero di riga e la y  uguale al numero di colonna

"""
Le immagini png sono gestite in rgba e non rgb, dove "a" sta per la trasparenza
"""

fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(imm1)
plt.show()
```

```python
"""
Per fare pi grafici in una figura, la si divide in una griglia.
"""

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(6,3))
# questo axs  un'array di assi
axs[0].scatter(x,y)
axs[0].set_title("Grafico a sinistra")
axs[1].imshow(imm1)
axs[1].set_title("Grafico a destra")

plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
# questo axs  un'array di assi
axs[0,0].scatter(x,y)
axs[0,0].set_title("Grafico sinistra alto")
axs[0,1].imshow(imm1)
axs[0,1].set_title("Grafico destra alto")
axs[1,0].imshow(imm)
axs[1,0].set_title("Grafico sinistra basso")
axs[1,1].imshow(matrice)
axs[1,1].set_title("Grafico destra basso")

plt.show()
```
