```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
matrice = np.random.uniform(0,255,(100,100))
matrice = matrice.astype(int)
matrice
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
axs[0].imshow(matrice, cmap="spring")
axs[1].hist(matrice.ravel(), bins=100) # con bins decido il numero di barre
plt.show()
```
