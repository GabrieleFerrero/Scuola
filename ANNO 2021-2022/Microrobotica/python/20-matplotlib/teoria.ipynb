### COM'È STRTTURATO UN GRAFICO DI MATPLOTLIB:


</br>
</br>

La base di matplotlib è la **figura**.
Gli assi sono la regione occupata dal grafico.
I tick sono le tacchette dove stanno i valori delle asciesse e delle ordinate.
Poi ci sono oggetti che stanno dentro il grafico vero e proprio:
- linea
- punti
- griglie

Poi c'è la legenda e serve in grafico in cui ci sono più punti e linee.

In matplotlib è tutto personalizzabile, ma bisogna scrivere molto codice.

```python
# es 1

fig, ax = plt.subplots(figsize=(10,10)) 
# così abbiamo creto un grafico con degli assi
plt.show()
```

```python
# es 2

fig, ax = plt.subplots(figsize=(10,10)) 
# ax.opzioni
plt.savefig("./nomefile.png") 
"""
Server per salvare il grafico in un file.
savefig capisce l'estensione il tipo di file
Estensioni supportate:
- jpeg
- jpg
- pdf
- png
- eps (formato vettoriale)
"""
```

### I grafici permettono di capire se i dati su cui si sta lavorando hanno una struttura oppure no.



```python
# es 3

x = np.linspace(0, 2*np.pi, 30)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(10,4))
ax.plot(x,y, "bo")
# b = blu
# o = punto
# tutte opzioni: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html?highlight=plot#matplotlib.pyplot.plot
plt.show()
```

```python
# es 4
n = 30
x = np.linspace(0, 2*np.pi, n)
y = np.sin(x)

z = np.random.normal(0,1, n)

fig, ax = plt.subplots(figsize=(10,4))
ax.plot(x,y, "b--", linewidth=5)
ax.plot(x,z, "r")
plt.show()
```

```python
# es 5

n = 100
x = np.linspace(0, 1, n)
y1 = x*0
y2 = x*1/2
y3 = x*1
y4 = x*2
y5 = x*4

fig, ax = plt.subplots(figsize=(20,8))
ax.plot(x,y1, "b--")
ax.plot(x,y2, "r:")
ax.plot(x,y3, "m|")
ax.plot(x,y4, "y_")
ax.plot(x,y5, "k3")
plt.show()
```

```python
# es 6

m_lista = np.array([0, 1/2, 1, 2, 4])
x = np.linspace(0, 1, 30)
lista_stili = ["g--", "b-", "yo", "r:", ".-"]

fig, ax = plt.subplots(figsize=(10,4))
for m,stile in zip(m_lista, lista_stili):
    ax.plot(x,x*m, stile)
plt.show()
```
