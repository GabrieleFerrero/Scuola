Scoprire il cliente di H&M che ha acquistato il set di articoli dal colore piÃ¹ vario (ogni utente ha acqusitato dei vestiti, e guardiamo la lista di colori diversi piÃ¹ lunga)

```python
lista_clienti = acquisti["customer_id"].unique()
lista_clienti
```

```python
colori_clienti = {"id_cliente":[], "numero_colori_differenti":[], "numero_prodotti_acquistati":[]}

# for cliente in lista_clienti:

for i in range(0,100): # da eliminare 
    cliente = lista_clienti[i] # da eliminare
    codici_prodotti_acquistati = acquisti[acquisti["customer_id"] == cliente]["article_id"]
    colori_prodotti_acquistati = articoli[articoli["article_id"].isin(codici_prodotti_acquistati)]["colour_group_name"]
    colori_clienti["id_cliente"].append(cliente)
    colori_clienti["numero_colori_differenti"].append(len(colori_prodotti_acquistati.unique()))
    colori_clienti["numero_prodotti_acquistati"].append(len(codici_prodotti_acquistati))
    
```

```python
colori_clienti
```

```python
col_clienti_df = pd.DataFrame(colori_clienti)
col_clienti_df
```

```python
n_max = col_clienti_df["numero_colori_differenti"].max() # guardo il numero di colori piÃ¹ grande
lista_persone_n_max = col_clienti_df[col_clienti_df["numero_colori_differenti"] == n_max]["id_cliente"] # guardo il numero di persone che hanno il numero di colori piÃ¹ grande

numero_prodotti_acquistati_piu_piccoli = col_clienti_df[(col_clienti_df["numero_colori_differenti"] == n_max) & (col_clienti_df["id_cliente"].isin(lista_persone_n_max))]["numero_prodotti_acquistati"].min() # guardo il numero piÃ¹ piccolo di vestiti comprati da queste persone

col_clienti_df[(col_clienti_df["id_cliente"].isin(lista_persone_n_max)) & (col_clienti_df["numero_colori_differenti"] == n_max) & (col_clienti_df["numero_prodotti_acquistati"] == numero_prodotti_acquistati_piu_piccoli)]["id_cliente"] # faccio la ricerca con i parametri ottenuti
```
