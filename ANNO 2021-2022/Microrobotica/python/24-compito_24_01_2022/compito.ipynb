```python
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
import numpy as np
import random
import pandas as pd
```

```python
mnist = fetch_openml("mnist_784")
```

```python
cifre = mnist["data"].values
cifre
```

```python
"""
1) Rappresentare 16 cifre scelte a caso nel dataset MNIST all'interno di una stessa figura matplotlib.
"""

fig, ax = plt.subplots(nrows=4, ncols=4, figsize=(15,15))
for y in range(4):
    for x in range(4):
        ax[x,y].set_title(f"{x} , {y}")
        ax[x,y].imshow(cifre[random.randint(0,len(cifre)-1)].reshape((28,28)), cmap="Greys")
plt.show()
```

```python
"""
2)Cerca su Internet degli open data sulla serie storica annuale della popolazione mondiale degli 
ultimi 50 anni.Scarica i dati in un file .csv e crea un grafico avente come ascissa 
lanno e come ordinata la popolazione.
"""

dati = pd.read_csv("./absolute-increase-global-population.csv")

anni = dati["Year"].values[-177:-77]
n_popolazione = dati["Absolute increase in population (OWID based on HYDE & UN)"].values[-177:-77]

fig, ax = plt.subplots(figsize =(10,10))
ax.plot(anni, n_popolazione)
plt.show()
```
